```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.signal import butter, lfilter,hilbert
import json
%load_ext autoreload
%autoreload 2
```

```python
data = np.load("hvmux_tests/20241109a-acquisitions_b.mpz",allow_pickle=True)["acquisitions"]
plt.figure(figsize=(15,7))
for i in range(len(data)):
    k = data[i][0]
    PAIRS = str(format(int(k), '#012b') )
    #print(PAIRS)
    if k > 0:
        POS = [pos for pos, char in enumerate(PAIRS) if char == '1'] 
        if not POS[0]%2:
            tx = (POS[0]+2)/2
            rx = (POS[1]+1)/2
        else:
            tx = (POS[1]+2)/2
            rx = (POS[0]+1)/2
        plt.plot(data[i][2900:6100],label=str(data[i][0])+": "+str(tx)+"/"+str(rx)+".")
plt.legend()
```
